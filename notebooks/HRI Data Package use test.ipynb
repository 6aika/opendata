# Jupyter notebook example on how to use data packages

Accompanied by [this blog post](https://dev.hel.fi/blog/restful-data)


```python
import pandas, datapackage
%matplotlib inline
```

```python
dp = datapackage.push_datapackage("datapackage.json", backend='pandas')
```

```python
dp.tables
```

```python
type(dp.tables)
```

```python
df = dp['data___data']
```

```python
df
```

```python
grouped = df.groupby('Address')
```

```python
grouped.first()
```

```python
grouped.describe()
```

```python
aggregated = grouped.count().sort_values(by='State of ticket payment', ascending=False)
```

```python
aggregated
```

```python
aggregated['State of ticket payment']
```

```python
aggregated[1:11]['State of ticket payment'].plot(kind='barh')
```
